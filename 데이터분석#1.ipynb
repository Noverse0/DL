```python
import pandas as pd
import numpy
eig = numpy.linalg.eig
```

```python
Data = pd.read_excel('Data.xlsx', index_col = 0)
Data
```

```python
Data.info()
```

```python
# Log Return
index_Data = (Data/Data.iloc[46,:])*100
#index_Data
log_retrun_Data = numpy.log(index_Data).diff()
log_retrun_Data = log_retrun_Data.dropna()
log_retrun_Data
```

```python
# 그냥 변화율
index_Data = (Data/Data.iloc[46,:])*100
#index_Data
log_retrun_Data = index_Data.diff()
log_retrun_Data = log_retrun_Data.dropna()
log_retrun_Data
```

```python
all_Data = log_retrun_Data.iloc[:,[0,3,4,5,6,7,8]]
capital_Data = log_retrun_Data.iloc[:,[1,3,4,5,6,7,8]]
region_Data = log_retrun_Data.iloc[:,[2,3,4,5,6,7,8]]
```

# 전국

## Correlation 구하기

```python
corr_data = all_Data.iloc[0:12].corr().iloc[[0],:]
for i in range(1,34):
    corr_data = corr_data.append(all_Data.iloc[i:12+i].corr().iloc[[0],:])

corr_data.index = all_Data.index[12:46]
corr_data = corr_data.drop(['전국_매매가격지수'], axis=1)
corr_data
```

## Eigenvalue 구하기

```python
eigen_Data = numpy.transpose(pd.DataFrame(eig(all_Data.iloc[0:12].corr())[0]))
for i in range(1,34):
    eigen_Data = eigen_Data.append(numpy.transpose(pd.DataFrame(eig(all_Data.iloc[0+i:12+i].corr())[0])))
eigen_Data.index = all_Data.index[12:46]
eigen_Data
```

## Absorption Raio 구하기

```python
ar_Data = eigen_Data
ar_Data[0]=ar_Data[0]/7
ar_Data[1]=(ar_Data[0]+ar_Data[1])/7
ar_Data[2]=(ar_Data[0]+ar_Data[1]+ar_Data[2])/7
ar_Data[3]=(ar_Data[0]+ar_Data[1]+ar_Data[2]+ar_Data[3])/7
ar_Data[4]=(ar_Data[0]+ar_Data[1]+ar_Data[2]+ar_Data[3]+ar_Data[4])/7
ar_Data[5]=(ar_Data[0]+ar_Data[1]+ar_Data[2]+ar_Data[3]+ar_Data[4]+ar_Data[5])/7
ar_Data[6]=(ar_Data[0]+ar_Data[1]+ar_Data[2]+ar_Data[3]+ar_Data[4]+ar_Data[5]+ar_Data[6])/7
```

```python
corr_data.to_excel('전국 상관관계 3년.xlsx')
eigen_Data.to_excel('전국 eigenvalue 3년.xlsx')
ar_Data.to_excel('전국 absorption ratio 3년.xlsx')
```
