```python
import numpy as np
import tensorflow as tf
```

## Simple Regression

먼저 가중치 W와 편향 b로 이루어져있는 Simple Regression을 만들어 보겠습니다.

```python
# simple-regression
x_train = np.array([1,2,3,4]) # Simple Regression이기 때문에 1차원의 벡터가 입력값으로 설정이 되었다.
y_train = np.array([0,-1,-2,-3])
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Input(shape=1)) #우리가 정의한 Model이 Input 값을 하나만 받는다는 뜻이다. 만약 2개를 받으면 Shape을 2로 바꾸어준다.
model.add(tf.keras.layers.Dense(units=1, activation='linear')) # unit은 출력값이 1개라는 의미이다.
```

```python
sgd = tf.keras.optimizers.SGD(learning_rate=0.01) # SGD는 확률적 경사하강법을 의미한다. 셈플을 뽑아 경사하강법을 진행하는 것이다.
```

```python
model.compile(loss='mse', optimizer=sgd) # Loss Function은 mse로 하고 Optimizer는 SGD로 설정하여 가중치를 업데이트한다.
```

```python
model.fit(x_train, y_train, epochs=100) # Model을 100번 학습시킨다. 출력된 Loss를보면 학습이 될수록 Loss 값이 줄어드는 것이 보인다.
```

```python
#Simple-Regression
model.predict(np.array([1,2,3,4])) # 학습된 Model로 결과를 확인해본다.
```

## Mutiple-Regression

```python
# mult-regression
x_train = np.array([[1,2],[2,3],[3,4],[4,5]])
y_train = np.array([0,-1,-2,-3])
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Input(shape=2))
model.add(tf.keras.layers.Dense(units=1, activation='linear'))
```

```python
sgd = tf.keras.optimizers.SGD(learning_rate=0.01)
```

```python
model.compile(loss='mse', optimizer=sgd)
```

```python
model.fit(x_train, y_train, epochs=100)
```

```python
#Multi-Regression
model.predict(np.array([[1,2],[2,3],[3,4],[4,5]]))
```
